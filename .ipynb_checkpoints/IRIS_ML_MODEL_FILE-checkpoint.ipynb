# Data Preprocessing

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
X = df.drop(['Species'],axis=1)
```

```python
X
```

```python
y=df["Species"]
```

```python
scaler = ss.fit(X)
```

```python
X_stdscaler = scaler.transform(X)
```

### Label Encoder

```python
y
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

### Split the DataSet for Training and Testing Purpose

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

## Model Selection

### Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn import metrics 
```

```python
svc = SVC(kernel="rbf")
svc.fit(X_train,y_train)
svc_pred = svc.predict(X_test)
svc_acc=metrics.accuracy_score(svc_pred,y_test)*100,'%'
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)
lr_acc=metrics.accuracy_score(lr_pred,y_test)*100,'%'
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_reg_pred = lr.predict(X_test)
lin_reg_acc=metrics.accuracy_score(lin_reg_pred,y_test)*100,'%'
```

### K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
knn_pred = knn.predict(X_test)
knn_acc=metrics.accuracy_score(knn_pred,y_test)*100,'%'
```

### Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy',random_state=7)
dtc.fit(X_train,y_train)
dtc_pred = dtc.predict(X_test)
dtc_acc=metrics.accuracy_score(dtc_pred,y_test)*100,'%'
```

### Gaussian Classification

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
gnb_pred = gnb.predict(X_test)
gnb_acc=metrics.accuracy_score(gnb_pred,y_test)*100,'%'
```
